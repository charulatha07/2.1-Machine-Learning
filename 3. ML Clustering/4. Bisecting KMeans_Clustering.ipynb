```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import BisectingKMeans

bisect_means = BisectingKMeans(n_clusters=5, init='k-means++', random_state=42)
y_bisect_means = bisect_means.fit_predict(X)
```

```python
y_bisect_means
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_bisect_means
```

```python
supervised
```

```python
supervised.to_csv("Bisecting_KMeans_Clusters.csv",index=False)
```

```python
centroids = bisect_means.cluster_centers_
```
