```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import Birch
birch = Birch(n_clusters=5)
y_birch = birch.fit_predict(X)
```

```python
y_birch
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_birch
```

```python
supervised
```

```python
supervised.to_csv("birch_clusters.csv",index=False)
```
