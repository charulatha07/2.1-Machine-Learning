```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import MeanShift
from sklearn.cluster import KMeans

mean_shift = MeanShift(bandwidth=10) 
y_meanshift = mean_shift.fit_predict(X)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans_limited = kmeans.fit_predict(mean_shift.cluster_centers_[y_meanshift])
```

```python
 y_kmeans_limited
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']= y_kmeans_limited
```

```python
supervised
```

```python
supervised.to_csv("meanshift_cluster.csv",index=False)
```
