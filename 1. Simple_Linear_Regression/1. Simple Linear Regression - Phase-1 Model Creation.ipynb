```python
#pandas is a library and it contains functions related to database
import pandas as pd
```

```python
#we are assigning the csv file to dataset variable
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#print the dataset
dataset
```

```python
# Split input and output

#assigning input to independent variabe by taking single column from dataset
independent=dataset[["YearsExperience"]]
#print the single column named YearsExperience
independent
```

```python
#assigning output to dependent variabe by taking single column from dataset
dependent=dataset[["Salary"]]
#print single column named Salary
dependent
```

```python
# Split training and test set

#sklearn is the library for Machine Learning algorithm
#from model_selection file we are importing train_test_split function
from sklearn.model_selection import train_test_split
#assigning 4 variables to get the output
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Model Creation 

#sklearn is the library for Machine Learning algorithm
#from linear_model file we are importing LinearRegression algorithm
from sklearn.linear_model import LinearRegression
#assigning LinearRegression() class to regressor variable
regressor=LinearRegression()
#by calling the fit function we are passing the input & output of training set
regressor.fit(x_train,y_train)
```

```python
#in coef_ we have slop value
weight=regressor.coef_
#print the value of slop
weight
```

```python
#in intercept_ we have origin or bais value
bais=regressor.intercept_
#print the value of origin
bais
```

```python
# Model Evaluation

#predicted input is stored in y_pred
y_pred=regressor.predict(x_test)
```

```python
#sklearn is the library for Machine Learning algorithm
#from metrics file we are importing r2_score
from sklearn.metrics import r2_score
#we are passing actual and predicted value to r_score
r_score=r2_score(y_test,y_pred)
#print
r_score
```

```python
#Save the Model

#we use pickle library to save the model
import pickle
#create file name .sav is the extension of pickle
filename="finalized_model_linear.sav"
#assigning the model by dump and our model is saved in regressor 
#wb is write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load and Check

#loading the file by using load keyword and rb is read binary
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#predicting the output for 15 years 
result=loaded_model.predict([[15]])
#print the output
result
```
